```python
#Conferindo as colunas

df.info() # Exibe um resumo conciso do DataFrame, incluindo tipos de dados e valores não nulos
df.describe() # Gera estatísticas descritivas das colunas numéricas do DataFrame
```

```python
# Seleciona características relevantes para a análise de 'Consumo inteligente'.
ici_features=df[[
    "Recency",
    "Frequency",
    "Monetary",
    "AvgInvoiceTicket"
]].copy() # Cria uma cópia para evitar modificar o DataFrame original 'df'

# Preenche quaisquer valores ausentes (NaN) nas características selecionadas com 0.
ici_features=ici_features.fillna(0)
# Exibe as primeiras 5 linhas do novo DataFrame 'ici_features' para verificar a seleção e a operação fillna.
ici_features.head()
```

```python
# Inicializa o MinMaxScaler, que irá escalar as características para um intervalo entre 0 e 1.
scaler = MinMaxScaler()

# Aplica o escalonamento nas 'ici_features' e cria um novo DataFrame 'normalized'.
# Os nomes das colunas são definidos para refletir que os dados foram normalizados.
normalized = pd.DataFrame(
    scaler.fit_transform(ici_features),
    columns=[
        "Recency_norm",          # Recência normalizada
        "Frequency_norm",        # Frequência normalizada
        "Monetary_norm",         # Valor monetário normalizado
        "AvgInvoiceTicket_norm"  # Ticket médio normalizado
    ]
)

# Exibe as primeiras 5 linhas do DataFrame 'normalized' para verificar os dados escalonados.
normalized.head()
```

```python
#Criando ICI
#Recency menor,Frequencia equilibrada : MELHOR
#Monetary alto:Alto consumo
#Ticket alto: Risco de gasto elevado

df["ICI"]= (
    (1-normalized["Recency_norm"]) *0.30 +
    (1-normalized["AvgInvoiceTicket_norm"]) *0.25 +
    (1-normalized["Monetary_norm"]) *0.25 +
     normalized["Frequency_norm"] *0.20
) *100

df["ICI"]=df["ICI"].round(2)
df[["CustomerID","Cluster","ICI"]].head()
```

```python
#Recomendações

def gerar_recomendacoes(row):

  if row["ICI"]>=80:
    return "Consumo eficiente. Manter o padrão atual e acompanhar evolução mensal."
  elif row ["ICI"]>=60:
    return "Bom padrão de consumo. Monitorar categorias com maior gasto para evitar desperdícios."
  elif row ["ICI"]>=40:
    return "Consumo moderado. Revisar frequência de compras ticket médio e categorias de maior impacto."
  else:
    return "Consumo crítico. Priorizar revisão de gastos, reduzir compras fora do padrão  e reavaliar categorias dominantes."

df["Recomendação"]=df.apply(gerar_recomendacoes,axis=1)
df[["CustomerID","ICI","Recomendação"]].head()
```

```python
#Distribuição do ICI
df["Recomendação"].value_counts()
```

```python
#Média ICI por cluster

df.groupby("Cluster")[["ICI","Recency","Frequency","Monetary","AvgInvoiceTicket"]].mean()
```

```python
#Média de preço por cluster

cluster_price_avg=df.groupby("Cluster")["AvgInvoiceTicket"].mean().reset_index()
cluster_price_avg.columns=["Cluster","ClusterAvgPrice"]
cluster_price_avg.head( )
```

```python
df=df.merge(cluster_price_avg,on="Cluster",how="left")
df.head()
```

```python
df["PriceIndex"]=df["AvgUnitPrice"]/df["ClusterAvgPrice"]

df[["CustomerID","Cluster","AvgUnitPrice","ClusterAvgPrice","PriceIndex"]].head()
```

```python
df.rename(columns={"Recomendação":"Recomendacao"},inplace=True)
#df.rename(columns={"AvgUnitPrice":"MeanProductPrice"})
```
