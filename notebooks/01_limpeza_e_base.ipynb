```python
#Carregando Dataset
import pandas as pd
df = pd.read_excel("/content/drive/MyDrive/Consumo inteligente hackathon/data/raw _dados originais /Online_Retail.xlsx",engine="openpyxl")
df.head()
```

```python
#Dimensão do Dataset
print("Linhas e Colunas:",df.shape)
```

```python
# Informações gerais
df.info( )
```

```python
#Verificando valores nulos
df.isnull().sum()
```

```python
#Estatísticas Basicas
df.describe()
```

```python
#Verificação de duplicados
print("Duplicados",df.duplicated().sum())
```

```python
#Limpeza inicial
df_clean=df.copy( )

df_clean=df_clean.dropna(subset=["CustomerID"] )
df_clean=df_clean[df_clean["Quantity"]>0]
df_clean=df_clean[df_clean["UnitPrice"]>0]

print("Dataset após limpeza:",df_clean.shape)
df_clean.head()
```

```python
#Converte data
df_clean["InvoiceDate"]=pd.to_datetime(df_clean["InvoiceDate"])
df_clean.info()
```

```python
#Criando Coluna de valor total
df_clean["TotalValue"]=df_clean["Quantity"]*df_clean["UnitPrice"]
df_clean.head()
```

```python
#Criando colunas auxiliares

df_clean["Month"]=df_clean["InvoiceDate"].dt.to_period("M").astype(str)
df_clean["DayOfWeek"]=df_clean["InvoiceDate"].dt.day_name()
df_clean["Hour"]=df_clean["InvoiceDate"].dt.hour
df_clean["Day"]=df_clean["InvoiceDate"].dt.day

df_clean.head( )
```

```python
#Resultado final
df_clean.info()
df_clean.describe( )
```

```python
#Salvando dataset tratado em CSV no Google Drive
output_path = '/content/drive/MyDrive/Consumo inteligente hackathon/data/processed dados tratados/Online_Retail_clean.csv'
df_clean.to_csv(output_path, index=False)
print(f"Arquivo salvo em: {output_path}")
```
