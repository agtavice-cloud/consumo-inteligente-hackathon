```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
```

```python
#Carregando customer_features.csv

customer_features = pd.read_csv("/content/drive/MyDrive/Consumo inteligente hackathon/data/processed_dados tratados/customer_features.csv")
customer_features.head()
```

```python
#Validação

customer_features.shape
```

```python
# Importando bibliotecas para clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Selecionando e preparando as features para clustering
features=customer_features[["Recency","Frequency","Monetary","AvgInvoiceTicket"]].copy()
features=features.fillna(0)

features.head()
```

```python
#Padronizando os dados

scaler = StandardScaler()
features_scaled= scaler.fit_transform(features)
```

```python
# Aplicando o algoritmo KMeans para clustering e atribuindo os resultados
kmeans= KMeans(n_clusters=4,random_state=42)
customer_features["Cluster"]=kmeans.fit_predict(features_scaled)

customer_features.head()
```

```python
#Quantidade de clientes por cluster

customer_features["Cluster"].value_counts()
```

```python
# Criando o perfil de cada cluster calculando a média das features
cluster_profile=customer_features.groupby("Cluster")[["Recency","Frequency","Monetary","AvgInvoiceTicket"]].mean()
cluster_profile
```

```python
# Salvando os dados do cliente com os clusters atribuídos em um novo arquivo CSV
customer_features.to_csv(
    "/content/drive/MyDrive/Consumo inteligente hackathon/data/processed_dados tratados/customer_features_cluster.csv",
    index=False,
)

print("Arquivo customer_features_cluster.csv salvo no Drive!")
```
